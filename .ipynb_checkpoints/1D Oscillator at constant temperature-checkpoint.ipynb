```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from mpl_toolkits.mplot3d import Axes3D
```

```python
# System parameters

kB = 1.38e-23     # Boltzman Constant
m = 18 * 1.67e-27 # Mass
k = 1e-18         # Spring constant
T = 300.0         # Target temperature (kT)
Q = 1e-15         # Thermostat coupling constant
dt = 0.01         # Time step
num_steps = 10000
```

```python
# Initial conditions
x = 0
p = np.sqrt(0.2)
alpha = 0.0  # Initial thermostat variable
```

```python
# Initialize arrays
X = np.zeros(num_steps)       # Position
P = np.zeros(num_steps)       # Momentum
E = np.zeros(num_steps)       # Total energy
```

```python
for i in range(num_steps):
    X[i] = x
    P[i] = p
    
    alpha += (1 / Q) * (p**2/m - kB * T) * dt
    x += p/m * dt
    p += -(k * X[-1] * dt) - (alpha * p * dt)
```

```python
plt.title('Phase space trajectory for 1D Harmonic Oscillator at constant Temperature')
plt.xlabel('Position x')
plt.ylabel('Momentum p')
plt.plot(X,P)
```

## Probability distribution plot

```python
# Compute KDE for momentum p
kde_p = gaussian_kde(P)
p_range = np.linspace(min(P), max(P), 100)
P_p = kde_p(p_range)

# Plot P(x)
plt.figure(figsize=(8, 4))
plt.plot(p_range * 1e10, P_p, color="blue", label=r"$P(x)$")
plt.xlabel("Position x (nm)")
plt.ylabel("Probability Density")
plt.title("Probability Distribution of Momentum P(p)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Compute KDE for position x
kde_x = gaussian_kde(X)
x_range = np.linspace(min(X), max(X), 100)
P_x = kde_x(x_range)

# Plot P(x)
plt.figure(figsize=(8, 4))
plt.plot(x_range * 1e10, P_x, color="blue", label=r"$P(x)$")
plt.xlabel("Position x (nm)")
plt.ylabel("Probability Density")
plt.title("Probability Distribution of Position P(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Define phase space grid for plotting
x_range = np.linspace(min(X), max(X), 100)
p_range = np.linspace(min(P), max(P), 100)
X_grid, P_grid = np.meshgrid(x_range, p_range)

# Compute 2D probability density using Gaussian KDE
data = np.vstack([X, P])
kde = gaussian_kde(data)
density = kde(np.vstack([X_grid.ravel(), P_grid.ravel()])).reshape(X_grid.shape)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

ax.plot_surface(X_grid, P_grid, density, cmap="viridis", edgecolor="k", alpha=0.8)

ax.set_xlabel("Position x (nm)")
ax.set_ylabel("Momentum p (kg·m/s)")
ax.set_zlabel("P(x, p)")
ax.set_title("3D Phase Space Probability Density P(x, p)")

plt.show()
```
