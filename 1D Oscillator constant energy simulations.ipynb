# One dimensional Harmonic Oscillator at constant energy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E = 10
m = 1
k = 1
dt = 0.01
dx = 0.01
```

## Phase space trajectory

```python
X = []
P = []

def calculate_p(x):
    return np.sqrt(2 * m * (E - 0.5 * k *( x**2)))

x = 0
p = calculate_p(x)


for i in range(0, 630):
    X.append(x)
    P.append(p)
    x += p/m * dt
    p += -k * x * dt
```

```python
plt.title('Phase space trajectory for 1D Harmonic Oscillator at constant Energy')
plt.xlabel('Position x')
plt.ylabel('Momentum y')
plt.plot(X, P)
plt.show()
```

## Number of accessible microstates

## Probability distribution plot

```python
hist, xedges, yedges = np.histogram2d(X, P, bins=200, density=True)
X_grid, Y_grid = np.meshgrid(xedges[:-1], yedges[:-1])

# Create plots
fig, axs = plt.subplots(1, 1, figsize=(6, 5))
pcm = axs.pcolormesh(X_grid, Y_grid, hist.T, cmap="plasma", shading='auto')

axs.set_xlabel("Position (x)")
axs.set_ylabel("Momentum (p)")
axs.set_title("Probability Distribution in Phase Space")
fig.colorbar(pcm, label="Probability Density")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Create surface plot
ax.plot_surface(X_grid, Y_grid, hist.T, cmap="plasma", edgecolor='none', alpha=0.9)

# Labels and title
ax.set_xlabel("Position (x)")
ax.set_ylabel("Momentum (p)")
ax.set_zlabel("Probability Density")
ax.set_title("3D Probability Density in Phase Space")

plt.show()
```
