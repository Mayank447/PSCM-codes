```python
import numpy as np
import matplotlib.pyplot as plt
from biopandas.pdb import PandasPdb
```

## Extracting the atom coordinates from pdb file

```python
# Reinitialize variables for frame extraction
frames = []
current_frame = []

with open("../case-study-water/section06/300K_coord.pdb", "r") as file:
    for line in file:
        if line.startswith(("ATOM", "HETATM")):
            x = float(line[30:38].strip())
            y = float(line[38:46].strip())
            z = float(line[46:54].strip())
            current_frame.append([x, y, z])

        elif line.startswith("END"):  # Marks end of a frame
            if current_frame:  # Store only if frame has data
                frames.append(np.array(current_frame))
                current_frame = []  # Reset for next frame

# Convert list of frames into a NumPy array
frames_np = np.array(frames, dtype=np.float32)

# Shape of the final array: (num_frames, num_atoms, 3)
frames_np.shape
```

## Position correlation

```python
def compute_position_vector_correlation(frames_np, max_tau=None):
    """
    Compute the correlation function for the full position vector r = (x, y, z) over different lag times τ.
    
    Args:
        frames_np (np.ndarray): Array of shape (num_frames, num_atoms, 3) containing atom positions.
        max_tau (int, optional): Maximum lag time τ. Defaults to num_frames // 2.
    
    Returns:
        np.ndarray: Correlation function C_r(τ) for each τ.
    """
    num_frames, num_atoms, _ = frames_np.shape
    if max_tau is None:
        max_tau = num_frames // 2  # Limit τ to avoid excessive noise

    correlation = np.zeros(max_tau, dtype=np.float32)
    correlation[0] = np.mean(np.sum(frames_np * frames_np, axis=2))
    
    for tau in range(1, max_tau):  # Start tau from 1 to avoid empty slices
        dot_product = np.sum(frames_np[:-tau] * frames_np[tau:], axis=2)  # Dot product (x*x + y*y + z*z)
        correlation[tau] = np.mean(dot_product)  # Average over all atoms and frames

    return correlation

# Compute the correlation function for the full position vector
correlation_r = compute_position_vector_correlation(frames_np)

plt.title('Correlation function in position')
plt.xlabel('Tau (time difference)')
plt.ylabel('C_r(tau)')
plt.plot(correlation_r)
```

## Velocity correlation

```python
def compute_velocities(frames_np, delta_t=1.0):
    """
    Compute velocities using finite difference approximation: 
    v_i(t) = (r_i(t+Δt) - r_i(t)) / Δt
    
    Args:
        frames_np (np.ndarray): Array of shape (num_frames, num_atoms, 3) containing atom positions.
        delta_t (float): Time step between frames.
    
    Returns:
        np.ndarray: Velocities of shape (num_frames-1, num_atoms, 3).
    """
    velocities = (frames_np[1:] - frames_np[:-1]) / delta_t
    return velocities
```

```python
def compute_velocity_correlation(velocities_np, max_tau=None):
    """
    Compute the velocity autocorrelation function (VACF).
    
    Args:
        velocities_np (np.ndarray): Array of shape (num_frames-1, num_atoms, 3) containing velocities.
        max_tau (int, optional): Maximum lag time τ. Defaults to num_frames // 2.
    
    Returns:
        np.ndarray: Velocity correlation function C_v(τ) for each τ.
    """
    num_frames, num_atoms, _ = velocities_np.shape
    if max_tau is None:
        max_tau = num_frames // 2  # Limit τ to avoid excessive noise

    correlation = np.zeros(max_tau, dtype=np.float32)
    correlation[0] = np.mean(np.sum(velocities_np * velocities_np, axis=2))

    for tau in range(1, max_tau):  # Start tau from 1 to avoid empty slices
        dot_product = np.sum(velocities_np[:-tau] * velocities_np[tau:], axis=2)  # v(t) · v(t+τ)
        correlation[tau] = np.mean(dot_product)  # Average over all atoms and frames

    return correlation

# Compute velocities
velocities_np = compute_velocities(frames_np)

# Compute velocity autocorrelation function (VACF)
correlation_v = compute_velocity_correlation(velocities_np)

plt.title('Correlation function in velocity')
plt.xlabel('Tau (time difference)')
plt.ylabel('C_v(tau)')
plt.plot(correlation_v)
```

## Force correlation

```python
# Lennard-Jones Parameters (assuming Argon-like system)
epsilon = 0.1554   # kcal/mol
sigma = 3.166  # Angstroms
cutoff = 10.0  # Cutoff distance for interactions (Angstroms)

def calculate_force(frames_np, epsilon=epsilon, sigma=sigma, cutoff=cutoff):
    """
    Uses Leonard Jones Potential. F = \grad(U).
    For force on a single particle, we average over force due to all other particles/
    """
    num_frames, num_atoms, _ = frames_np.shape
    forces = np.zeros_like(frames_np)

    for f in range(num_frames):
        frame = frames_np[f]
        force_frame = np.zeros((num_atoms, 3))

        # Compute pairwise displacement vectors
        disp_vectors = frame[:, np.newaxis, :] - frame[np.newaxis, :, :]  # Shape: (num_atoms, num_atoms, 3)
        
        # Compute pairwise distances
        distances = np.linalg.norm(disp_vectors, axis=2)  # Shape: (num_atoms, num_atoms)
        np.fill_diagonal(distances, np.inf)  # Avoid self-interaction
        
        # Apply cutoff mask
        within_cutoff = distances < cutoff
        r_inv = np.where(within_cutoff, 1.0 / distances, 0.0)  # Avoid division by zero
        r6 = (sigma * r_inv) ** 6
        r12 = r6 ** 2

        # Compute force magnitudes
        force_magnitudes = np.where(within_cutoff, 24 * epsilon * (2 * r12 - r6) * r_inv**2, 0.0)

        # Compute force vectors
        force_vectors = force_magnitudes[..., np.newaxis] * disp_vectors
        force_frame = np.sum(force_vectors, axis=1)  # Sum over all interacting pairs

        forces[f] = force_frame

    return forces
```

```python
def compute_force_correlation(forces_np, max_tau=None):
    """
    Compute the force autocorrelation function (FACF).
    
    Args:
        forces_np (np.ndarray): Array of shape (num_frames, num_atoms, 3) containing forces.
        max_tau (int, optional): Maximum lag time τ. Defaults to num_frames // 2.
    
    Returns:
        np.ndarray: Force correlation function C_F(τ) for each τ.
    """
    num_frames, num_atoms, _ = forces_np.shape
    if max_tau is None:
        max_tau = num_frames // 2  # Limit τ to avoid excessive noise

    correlation = np.zeros(max_tau, dtype=np.float32)
    correlation[0] = np.mean(np.sum(forces_np * forces_np, axis=2))

    for tau in range(1, max_tau):  # Start tau from 1 to avoid empty slices
        dot_product = np.sum(forces_np[:-tau] * forces_np[tau:], axis=2)  # F(t) · F(t+τ)
        correlation[tau] = np.mean(dot_product)  # Average over all atoms and frames

    return correlation

# Compute forces using optimized NumPy approach
forces = calculate_force(frames_np)
```

```python
# Compute FACF using NumPy-only forces
correlation_f = compute_force_correlation(forces)

plt.title('Correlation function in Force')
plt.xlabel('Tau (time difference)')
plt.ylabel('C_F(tau)')
plt.plot(correlation_f)
```
